```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#Date From 2018 January 01 to 2022 November 15
```

```python
start = datetime.datetime(2018,1,1)
end   = datetime.datetime(2022,11,15)
```

```python
google = web.DataReader("GOOGL",'yahoo',start,end)
tesla = web.DataReader("TSLA",'yahoo',start,end)
ford = web.DataReader("F",'yahoo',start,end)
```

```python
#Save all stcoks data files to csv
```

```python
google.to_csv("Google_stocks.csv")
ford.to_csv("Ford_Stocks.csv")
tesla.to_csv("Tesla_Stocks.csv")
```

```python
#Plotting stock Open Price for all stocks
```

```python
tesla["Open"].plot(label = "Tesla", figsize = (15,7))
google["Open"].plot(label = "Google")
ford["Open"].plot(label = "Ford")
plt.ylabel("Stock Price")
plt.title("Stock Price of Google, Tesal and Ford")
plt.legend()
```

```python
#Plotting volume of each stocks
```

```python
google["Volume"].plot(label = "Google", figsize = (15,7))
tesla["Volume"].plot(label = "Tesla")
ford["Volume"].plot(label = "Ford")
plt.ylabel("Volume Traded")
plt.legend()
```

```python
#Comparing the Total Traded for all stocks
```

```python
google["Total Traded"] = google["Open"] * google["Volume"]
tesla["Total Traded"] = google["Open"] * tesla["Volume"]
ford["Total Traded"] = google["Open"] * ford["Volume"]
```

```python
#Histogram visualization of the data
```

```python
google["returns"] = (google["Close"]/google["Close"].shift(1)) - 1
```

```python
google.head()
```

```python
tesla["returns"] = (tesla["Close"]/tesla["Close"].shift(1)) - 1
```

```python
tesla.head()
```

```python
ford["returns"] = (ford["Close"]/ford["Close"].shift(1)) - 1
```

```python
ford.head()
```

```python
google["returns"].hist()
```

```python
tesla["returns"].hist()
```

```python
ford["returns"].hist()
```

```python
google["returns"].hist(label = "Google",alpha = 0.6, figsize = (13,8))
tesla["returns"].hist(label = "Tesla",alpha = 0.6)
ford["returns"].hist(label = "Ford",alpha = 0.6)
plt.legend()
```

```python
#Filtering the histogram for the stocks
```

```python
google["returns"].plot(kind = "kde", label = "Google", figsize = (15,6))
tesla["returns"].plot(kind = "kde", label = "Tesla")
ford["returns"].plot(kind = "kde", label = "Ford")
plt.legend()
```

```python
#Box plot interpretation
```

```python
box_df = pd.concat([google["returns"],tesla["returns"],ford["returns"]],axis =1)
box_df.columns = ["Google Returns","Tesla Returns","Ford Returns"]
box_df.plot(kind = 'box',figsize = (16,6))
```

```python
scatter_matrix(box_df,figsize = (10,10),hist_kwds = {"bins": 50}, alpha = 0.25)
```

```python
#Cumulative Return
```

```python
google['Cumulative Return'] = (1 + google['returns']).cumprod()
```

```python
google.head()
```

```python
tesla['Cumulative Return'] = (1 + tesla['returns']).cumprod()
```

```python
tesla.head()
```

```python
ford['Cumulative Return'] = (1 + ford['returns']).cumprod()
```

```python
ford.head()
```

```python
google["Cumulative Return"].plot(label = "Google", figsize = (15,8))
tesla["Cumulative Return"].plot(label = "Tesla")
ford["Cumulative Return"].plot(label = "Ford")
plt.legend()
```
